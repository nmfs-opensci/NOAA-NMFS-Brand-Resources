```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
# inspired: 
# https://github.com/nmfs-fish-tools/nmfspalette/tree/main
# https://connect.fisheries.noaa.gov/colors/

# Define palettes grouped by themes
nmfs_palettes = {
    "oceans": ["#001743", "#002364", "#003087", "#0085CA", "#5EB6D9", "#C6E6F0"],
    "waves": ["#005E5E", "#00797F", "#1EBEC7", "#90DFE3"],
    "seagrass": ["#365E17", "#4B8320", "#76BC21", "#B1DC6B"],
    "urchin": ["#3B469A", "#5761C0", "#737BE6", "#A8B8FF"],
    "crustacean": ["#853B00", "#DB6015", "#FF8400", "#ffab38"],
    "coral": ["#901200", "#b71300", "#db2207", "#ff6c57"],
}



def display_palette(colors, n=100):
    """Generates and displays a gradient based on a list of colors."""
    # Create a color map
    cmap = LinearSegmentedColormap.from_list("Custom", colors, N=n)
    gradient = np.linspace(0, 1, 256)
    gradient = np.vstack((gradient, gradient))
    
    # Plotting
    fig, ax = plt.subplots(figsize=(6, 2))
    ax.imshow(gradient, aspect='auto', cmap=cmap)
    ax.set_axis_off()
    plt.show()

def display_nmfs_palette(name, n):
    """Displays the named palette with n interpolations."""
    if name in nmfs_palettes:
        display_palette(nmfs_palettes[name], n)

# Example usage
for theme in ["oceans", "waves", "seagrass", "urchin", "crustacean", "coral"]:
    display_nmfs_palette(theme, 10)
```

# Create Color Palettes w/ Hex Codes

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap, hex2color, to_hex

# Define NMFS palettes
nmfs_palettes = {
    "oceans": ["#001743", "#002364", "#003087", "#0085CA", "#5EB6D9", "#C6E6F0"],
    "waves": ["#005E5E", "#00797F", "#1EBEC7", "#90DFE3"],
    "seagrass": ["#365E17", "#4B8320", "#76BC21", "#B1DC6B"],
    "urchin": ["#3B469A", "#5761C0", "#737BE6", "#A8B8FF"],
    "crustacean": ["#853B00", "#DB6015", "#FF8400", "#ffab38"],
    "coral": ["#901200", "#b71300", "#db2207", "#ff6c57"]
}

# Define the palette names and corresponding save paths
palettes_to_save = {
    "oceans": "oceans_palette.png",
    "waves": "waves_palette.png",
    "seagrass": "seagrass_palette.png",
    "urchin": "urchin_palette.png",
    "crustacean": "crustacean_palette.png",
    "coral": "coral_palette.png"
}

def interpolate_colors(color_start, color_end, n=100):
    """
    Interpolates n colors between two hex colors, inclusive of endpoints.
    
    Parameters:
        color_start (str): Starting color in hex format.
        color_end (str): Ending color in hex format.
        n (int): Number of colors to interpolate.
    
    Returns:
        list: List of interpolated hex colors.
    """
    start_rgb = hex2color(color_start)
    end_rgb = hex2color(color_end)
    # Including endpoints
    return [to_hex((start_rgb[0] * (1 - t) + end_rgb[0] * t,
                    start_rgb[1] * (1 - t) + end_rgb[1] * t,
                    start_rgb[2] * (1 - t) + end_rgb[2] * t)) for t in np.linspace(0, 1, n)]

def display_palette_with_interpolations(palette_colors, num_colors=10, save_path=None):
    """
    Displays a gradient and annotates n evenly spaced hex codes for a given palette.
    
    Parameters:
        palette_colors (list): List of hex colors in the palette.
        num_colors (int): Number of hex codes to annotate.
        save_path (str): Path to save the displayed palette image.
    """
    total_interpolated_colors = []
    num_transitions = len(palette_colors) - 1
    total_colors = 0
    for i in range(num_transitions):
        # Calculate the number of colors to interpolate between each pair of colors
        colors_between = num_colors // num_transitions
        if i == num_transitions - 1:
            colors_between += num_colors % num_transitions  # Add remaining colors to the last transition
        # Interpolate colors
        interpolated_colors = interpolate_colors(palette_colors[i], palette_colors[i+1], colors_between + 1)
        total_interpolated_colors.extend(interpolated_colors[:-1])  # Exclude the last color to avoid duplication
        total_colors += colors_between

    # Ensure that the total number of colors is exactly num_colors
    if total_colors < num_colors:
        last_color = interpolate_colors(palette_colors[-2], palette_colors[-1], 2)[1]
        total_interpolated_colors.extend([last_color] * (num_colors - total_colors))

    # Create the gradient
    cmap = LinearSegmentedColormap.from_list("Custom", total_interpolated_colors, N=num_colors)
    gradient = np.linspace(0, 1, num_colors).reshape(1, num_colors)
    fig, ax = plt.subplots(figsize=(12, 1))
    img = ax.imshow(gradient, aspect='auto', cmap=cmap)
    ax.set_axis_off()

    # Annotate n hex codes evenly spaced along the gradient
    spacing = len(total_interpolated_colors) // num_colors
    for i in range(num_colors):
        pos_x = i * spacing
        pos_y = 0  # Center vertically in the bar
        hex_code = total_interpolated_colors[pos_x]
        ax.text(pos_x, pos_y, hex_code, color='white', ha='center', va='center', fontweight='bold')

    if save_path:
        plt.savefig(save_path, bbox_inches='tight', pad_inches=0)
    plt.show()

# Loop through each palette, interpolate colors, and save it
for palette_name, save_path in palettes_to_save.items():
    display_palette_with_interpolations(nmfs_palettes[palette_name], num_colors=10, save_path=save_path)
```

# Create Color Swatches

```python
from PIL import Image
import matplotlib.pyplot as plt

def create_color_swatch(hex_color, filename):
    # Create an image with RGB mode
    image = Image.new("RGB", (20, 20), hex_color)
    # Save the image
    image.save(filename)
    # Display the image with smaller size
    plt.figure(figsize=(0.5, 0.5))  # Adjust the figure size here
    plt.imshow(image)
    plt.axis('off')  # Turn off axis
    plt.show()

# Define all your colors grouped by category
nmfs_palettes = {
    "oceans": ["#001743", "#002364", "#003087", "#0085CA", "#5EB6D9", "#C6E6F0"],
    "waves": ["#005E5E", "#00797F", "#1EBEC7", "#90DFE3"],
    "seagrass": ["#365E17", "#4B8320", "#76BC21", "#B1DC6B"],
    "urchin": ["#3B469A", "#5761C0", "#737BE6", "#A8B8FF"],
    "crustacean": ["#853B00", "#DB6015", "#FF8400", "#ffab38"],
    "coral": ["#901200", "#b71300", "#db2207", "#ff6c57"],
}

# Generate and save each color swatch
for palette_name, palette_colors in nmfs_palettes.items():
    for index, hex_color in enumerate(palette_colors):
        filename = f"{palette_name}{index+1}.png"
        create_color_swatch(hex_color, filename)
```
